### SQL Queries

```python
from tabulate import tabulate
```

#### Medium level

```python
# 1. Total Number of Police Stops

mycursor.execute("SELECT count(*) AS No_of_Police_stops FROM police_check_data")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))
```

```python
# 2. Count of Stops by Violation Type

mycursor.execute("SELECT Violation, COUNT(*) AS Violation_stops FROM police_check_data GROUP BY Violation ORDER BY Violation_stops DESC")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))
```

```python
# 3. Number of Arrests vs. Warnings

mycursor.execute("SELECT stop_outcome,COUNT(*) AS Count FROM police_check_data WHERE stop_outcome IN ('Arrest', 'Warning') GROUP BY stop_outcome ORDER BY count DESC")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 4. Average Age of Drivers Stopped

mycursor.execute("SELECT AVG(driver_age) AS Avg_driver_age  FROM police_check_data")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 5. Top 5 Most Frequent Search Types

mycursor.execute("SELECT  search_type, COUNT(*) AS count   FROM police_check_data GROUP BY search_type ORDER BY count DESC LIMIT 5")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 6. Count of Stops by Gender

mycursor.execute("SELECT  driver_gender AS gender, COUNT(*) AS stop_count   FROM police_check_data GROUP BY driver_gender ")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 7. Most Common Violation for Arrests

mycursor.execute("SELECT  Violation, COUNT(*) AS arrest_count   FROM police_check_data WHERE is_arrested=TRUE GROUP BY Violation  ORDER BY arrest_count DESC LIMIT 1")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 8. Average Stop Duration for Each Violation

mycursor.execute("""SELECT  Violation, AVG( CASE 
WHEN stop_duration = '0-15 Min' THEN 7.5
    WHEN stop_duration = '16-30 Min' THEN 23
    WHEN stop_duration = '30+ Min' THEN 35
    ELSE NULL
  END) AS avg_stop_duration   FROM police_check_data GROUP BY Violation""")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 9. Number of Drug-Related Stops by Year

mycursor.execute("SELECT  YEAR(stop_date) AS stop_year, COUNT(*) AS drugs_related_stop_counts  FROM police_check_data WHERE drugs_related_stop=TRUE GROUP BY stop_year  ORDER BY stop_year")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 10. Drivers with the Highest Number of Stops

mycursor.execute("SELECT  vehicle_number, COUNT(*) AS drivers_stops  FROM police_check_data GROUP BY vehicle_number  ORDER BY drivers_stops DESC LIMIT 5")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 11. Number of Stops Conducted at Night (Between 10 PM - 5 AM)

mycursor.execute("SELECT   COUNT(*) AS stops_at_night  FROM police_check_data WHERE TIME(stop_time)>= '22:00:00' OR TIME(stop_time) <= '05:00:00'")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 12. Number of Searches Conducted by Violation Type

mycursor.execute("SELECT violation,  COUNT(*) AS count_of_searches  FROM police_check_data WHERE search_conducted=TRUE GROUP BY violation ORDER BY count_of_searches DESC")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 13. Arrest Rate by Driver Gender

mycursor.execute("SELECT driver_gender,  ROUND(SUM(CASE WHEN is_arrested = TRUE THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS arrest_rate  FROM police_check_data  GROUP BY driver_gender ")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 14. Violation Trends Over Time (Monthly Count of Violations)


mycursor.execute("SELECT violation,  MONTH(stop_date) AS month,  COUNT(*)  AS counts  FROM police_check_data  GROUP BY MONTH(stop_date), violation ORDER BY counts DESC")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 15. Most Common Stop Outcomes for Drug-Related Stops


mycursor.execute("SELECT stop_outcome,  COUNT(*)  AS counts  FROM police_check_data WHERE drugs_related_stop=TRUE GROUP BY stop_outcome ORDER BY counts DESC")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

#### Complex level

```python
# 1. Yearly Breakdown of Stops and Arrests by Country (Using Subquery and Window Functions)


mycursor.execute("""SELECT year, country_name, total_stops, arrest_count FROM (SELECT country_name,YEAR(stop_date) AS year, 
                 COUNT(*) OVER (PARTITION BY YEAR(stop_date), country_name) AS total_stops, 
                 SUM(CASE WHEN is_arrested = TRUE THEN 1 ELSE 0 END)
                  OVER (PARTITION BY YEAR(stop_date), country_name) AS arrest_count 
                  FROM police_check_data) AS sub
GROUP BY year, country_name, total_stops, arrest_count
ORDER BY year, country_name """)
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 2. Driver Violation Trends Based on Age and Race (Join with Subquery)


mycursor.execute("SELECT violation, sub.age, sub.race,   COUNT(*)  AS counts  FROM (SELECT  violation, driver_age AS age, driver_race AS race FROM police_check_data ) AS sub  GROUP BY violation,  sub.age, sub.race ORDER BY counts DESC  ")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 3. Time Period Analysis of Stops (Joining with Date Functions)


mycursor.execute("SELECT YEAR(STR_TO_DATE(stop_date, '%Y-%m-%d')) AS year, MONTH(STR_TO_DATE(stop_date, '%Y-%m-%d')) AS month, DAYNAME(STR_TO_DATE(stop_date, '%Y-%m-%d')) AS weekday, HOUR(STR_TO_DATE(stop_time, '%H:%i')) AS hour, COUNT(*)  AS stop_counts  FROM police_check_data GROUP BY  year, month, weekday, hour ORDER BY year, month, weekday, hour")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 4. Correlation Between Age, Violation, and Stop Duration (Subquery)


mycursor.execute(""" SELECT 
                 violation, 
                 driver_age_group,
                 ROUND(AVG(stop_duration_minutes), 2) AS avg_stop_duration,                  
                 COUNT(*)  AS counts 
                FROM 
                 (SELECT CASE WHEN driver_age < 18 THEN '<18'
            WHEN driver_age BETWEEN 18 AND 25 THEN '18-25'
            WHEN driver_age BETWEEN 26 AND 40 THEN '26-40'
            WHEN driver_age BETWEEN 41 AND 60 THEN '41-60'
            ELSE '60+'
        END AS driver_age_group,
        violation,
        CASE stop_duration
            WHEN '0-15 Min' THEN 10
            WHEN '16-30 Min' THEN 23
            WHEN '30+ Min' THEN 35
            ELSE NULL
        END AS stop_duration_minutes  FROM police_check_data) AS sub
                 GROUP BY violation, driver_age_group ORDER BY counts DESC""")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 5. Violations with High Search and Arrest Rates (Window Function)


mycursor.execute("SELECT violation, search_type, ROUND(AVG(search_conducted)*100, 2) AS search_rate, ROUND(AVG(is_arrested)*100, 2) AS arrest_rate, RANK() OVER (ORDER BY AVG(search_conducted) DESC) AS search_rank, RANK() OVER (ORDER BY AVG(is_arrested) DESC) AS arrest_rank,  COUNT(*)  AS counts  FROM police_check_data GROUP BY violation, search_type ORDER BY search_rank")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 6. Driver Demographics by Country (Age, Gender, and Race)


mycursor.execute("SELECT country_name,  driver_gender, ROUND(AVG(driver_age), 1) AS avg_age, driver_race, COUNT(*)  AS counts  FROM police_check_data  GROUP BY country_name, driver_gender, driver_race ORDER BY country_name, counts DESC")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```

```python
# 7.Top 5 Violations with Highest Arrest Rates


mycursor.execute("SELECT violation, ROUND(SUM(CASE WHEN is_arrested = TRUE THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS arrest_rate, COUNT(*)  AS police_stops  FROM police_check_data  GROUP BY violation ORDER BY arrest_rate DESC LIMIT 5")
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='grid'))

```
